# Uniform Cost Search

```python
from queue import PriorityQueue

def ucs(start, goal, G):
    expanded = [] # to store expanded nodes
    frontier = PriorityQueue()
    frontier.put((0,start,[start]))
    i = 1
    while not frontier.empty():
        node = frontier.get() # select a node for expansion
        u = node[1]
        if u in expanded:
          continue
        print()
        print(u+ " selected. Cost: "+ str(node[0]))
        if u == goal:# goal test
            print("*** GOAL TEST SATISFIED ***")
            print()
            return node # path
        expanded.append(u)
        print('['+str(i)+']: '+u+" expanding...")
        i = i +1
        for v in G[u].keys():#expanding
            if v not in expanded:
                pathcost = node[0] + G[u][v]
                path = node[2] + [v]
                frontier.put((pathcost,v,path))
                print(u + "->"+v +" : " + str(G[u][v]))
                #print("Path: "+ str(path) + ", Cost:"+ str(pathcost))
        #print(frontier.queue)
        for n in frontier.queue:
          print(n[1]+': '+ str(n[0]))
    print('Failed')
```

```python
ucs('Arad', 'Bucharest', romanian_map)
```

# Greedy Best First Search

```python
h_rm = {'Arad': 366, 'Zerind': 374, 'Oradea': 380, 'Sibiu': 253,
    'Timisoara': 329, 'Lugoj': 244, 'Mehadia': 241,
    'Drobeta': 242, 'Craiova': 160, 'Rimnicu': 193,
    'Fagaras': 176, 'Pitesti': 100, 'Bucharest': 0,
    'Giurgiu': 77, 'Urziceni': 80, 'Hirsova': 151,
    'Eforie': 161, 'Vaslui': 199, 'Iasi': 226, 'Neamt': 234}
```

```python
from queue import PriorityQueue

def gbfs(start, goal, G, h):
    expanded = [] # to store expanded nodes
    frontier = PriorityQueue()
    frontier.put((h[start],start,[start],0))
    i = 1
    while not frontier.empty():
        node = frontier.get() # select a node for expansion
        u = node[1]
        if u in expanded:
          continue
        print()
        print(u+ " selected. Cost: "+ str(node[0]))
        if u == goal:# goal test
            print("*** GOAL TEST SATISFIED ***")
            print()
            return node # path
        expanded.append(u)
        print('['+str(i)+']: '+u+" expanding...")
        i = i +1
        for v in G[u].keys():#expanding
            if v not in expanded:
                pathcost = node[3] + G[u][v]
                path = node[2] + [v]
                frontier.put((h[v],v,path,pathcost))
                print(u + "->"+v +" : " + str(G[u][v]))
        for n in frontier.queue:
          print(n[1]+': '+ str(n[0]))
    print('Failed')
```

```python
gbfs('Arad', 'Bucharest', romanian_map, h_rm) # change source only
```

# A*

```python
# code A* here
h_rm = {'Arad': 366, 'Zerind': 374, 'Oradea': 380, 'Sibiu': 253,
    'Timisoara': 329, 'Lugoj': 244, 'Mehadia': 241,
    'Drobeta': 242, 'Craiova': 160, 'Rimnicu': 193,
    'Fagaras': 176, 'Pitesti': 100, 'Bucharest': 0,
    'Giurgiu': 77, 'Urziceni': 80, 'Hirsova': 151,
    'Eforie': 161, 'Vaslui': 199, 'Iasi': 226, 'Neamt': 234}
```

```python
# Call A* here for the Romanian map
from queue import PriorityQueue

def A(start, goal, G, h):
    expanded = [] # to store expanded nodes
    frontier = PriorityQueue()
    frontier.put((h[start]+0,start,[start],0))
    i = 1
    while not frontier.empty():
        node = frontier.get() # select a node for expansion
        u = node[1]
        if u in expanded:
          continue
        print()
        print(u+ " selected. Cost: "+ str(node[0]))
        if u == goal:# goal test
            print("*** GOAL TEST SATISFIED ***")
            print()
            return node # path
        expanded.append(u)
        print('['+str(i)+']: '+u+" expanding...")
        i = i +1
        for v in G[u].keys():#expanding
            if v not in expanded:
                pathcost = node[3] + G[u][v]
                path = node[2] + [v]
                frontier.put((h[v]+pathcost,v,path,pathcost))
                print(u + "->"+v +" : " + str(G[u][v]))
        for n in frontier.queue:
          print(n[1]+': '+ str(n[0]))
    print('Failed')
```

```python
A('Arad', 'Bucharest', romanian_map, h_rm)
```
