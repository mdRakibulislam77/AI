# N Queen Attack Detection

```python
def drawBoard(P):
  N = len(P)
  for i in range(1,N+1):
    print("|", end="")
    for j in range(1,N+1):
      if (i,j) in P:
        ind = str(1+ P.index((i,j)))
        print(" "+ind+" |", end= "")
      else:
        print(" * |", end= "")
    print()
```

```python
import math
import random
def generate_positions(N):
    CL = list(range(0,N*N));
    P = []
    for i in range(N):
        p = random.choice(CL)
        CL.remove(p)
        r = math.ceil((p+1)/N)
        c = 1+p%N
        P.append((r,c))
       # print('Q{}: {},{}'.format(i+1,r,c))
    return P

generate_positions(4)
```

```python
generate_positions(4)
```

```python
N = 4
positions = generate_positions(N)
drawBoard(positions)
```

```python
# row attack
def rowAttacK(P):
  n = len(P)
  count = 0
  for i in range(n):
    for j in range(i+1,n):
      if P[i][0] == P[j][0]: #check row attack
        print('row attack:', end = " " )
        print(P[i], end = ", ")
        print(P[j])
        count = count + 1
  return count

rowAttacK(positions)
```

```python
#column attack
def columAttacK(P):
  n = len(P)
  count = 0
  for i in range(n):
    for j in range(i+1,n):
      if P[i][1] == P[j][1]: #check colum attack
        print('colum attack:', end = " " )
        print(P[i], end = ", ")
        print(P[j])
        count = count + 1
  return count

columAttacK(positions)
```

```python
# diagonal atack
def diagonalAttack(P):
  n = len(P)
  count = 0
  for i in range(n):
    for j in range(i+1,n):
      if abs(P[i][0] - P[j][0]) == abs(P[i][1] - P[j][1]):
        print('diagonal attack:', end = " " )
        print(P[i], end = ", ")
        print(P[j])
        count = count + 1
  return count

diagonalAttack(positions)
```

```python
# implement a method to print a state is valid or not

def isValid(P):
  n = len(P)
  for i in range(n):
    for j in range(i+1,n):
      if P[i][0] == P[j][0]:
        return False
      elif P[i][1] == P[j][1]:
        return False
      elif abs(P[i][0] - P[j][0]) == abs(P[i][1] - P[j][1]):
        return False

  return True

isValid(positions)
```

```python
# call isValid untill you get valid solution

while not isValid(positions):
    positions = generate_positions(N)

print("Valid configuration found:")
print(positions)
```
